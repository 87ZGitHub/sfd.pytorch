```python
%matplotlib inline

from dataset import create_wf_datasets, my_collate_fn
import torch
from config import Config
from utils import draw_bounding_boxes
```

```python
train_dataset, val_dataset = create_wf_datasets(Config.WF_DATASET_DIR)

train_dataloader = torch.utils.data.DataLoader(
    train_dataset,
    batch_size=Config.BATCH_SIZE,
    num_workers=Config.DATALOADER_WORKER_NUM,
    shuffle=True,
    collate_fn=my_collate_fn
)
```

```python
def show_index(index):
    data = train_dataset[index]
    for item in data:
        draw_bounding_boxes(item[0], item[1])
```

```python
for index, (images, all_gt_bboxes, path, _) in enumerate(train_dataloader):
    if index == 2:
        break
    
    images = images.numpy()
    
    for k in range(images.shape[0]):
        draw_bounding_boxes(images[k], all_gt_bboxes[k])
```
